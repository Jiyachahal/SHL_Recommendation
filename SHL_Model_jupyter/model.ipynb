```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    
    tokens = [stemmer.stem(word) for word in text.split() if word not in stop_words]
    return ' '.join(tokens)

df['Processed_Description'] = df['Description'].astype(str).apply(preprocess_text)

df['Processed_Description'] += " " + df['Job Levels'].astype(str).apply(preprocess_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python

def get_recommendations(user_input, df, vectorizer, X, top_n=10, min_similarity=0.1):
    
    processed_input = preprocess_text(user_input)
    user_vec = vectorizer.transform([processed_input])

    similarities = cosine_similarity(user_vec, X).flatten()

    sorted_indices = np.argsort(similarities)[::-1]

    filtered_indices = [i for i in sorted_indices if similarities[i] >= min_similarity]
    if not filtered_indices:
        filtered_indices = [sorted_indices[0]]  # Ensure at least one

    top_indices = filtered_indices[:top_n]
    top_scores = similarities[top_indices]

    return top_indices
```

```python
query = "Looking to hire mid-level professionals who are proficient in Python, SQL and JavaScript. Need an assessment package under 60 minutes."
top_indices = get_recommendations(query, df, vectorizer, X, top_n=10)
recommended_items =[top_indices]
recommended_items

```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def evaluate_model_with_function(df, vectorizer, X, top_k=3):
    recall_scores = []
    ap_scores = []

    descriptions = df['Processed_Description'].tolist()
    labels = df['Category_Code'].tolist()

    for idx, (desc, true_label) in enumerate(zip(descriptions, labels)):
        top_indices = get_recommendations(desc, df, vectorizer, X, top_n=top_k, min_similarity=0.1)

        top_indices = [i for i in top_indices if i != idx][:top_k]
        top_labels = [labels[i] for i in top_indices]

        relevant = sum([1 for label in top_labels if label == true_label])
        total_possible = sum([1 for i in range(len(labels)) if labels[i] == true_label and i != idx])
        recall = relevant / min(top_k, total_possible) if total_possible > 0 else 0
        recall_scores.append(recall)

        ap = 0
        num_hits = 0
        for i, label in enumerate(top_labels):
            if label == true_label:
                num_hits += 1
                ap += num_hits / (i + 1)
        ap_scores.append(ap)

    mean_recall_at_3 = np.mean(recall_scores)
    map_at_3 = np.mean(ap_scores)

    return mean_recall_at_3, map_at_3
```
